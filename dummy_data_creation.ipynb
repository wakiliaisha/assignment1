Problem 1

```python
import numpy as np
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
rand_num = np.random.multivariate_normal(mean, cov, 500)
print(rand_num)
rand_num.shape

```

Problem 4

```python
import numpy as np
mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
rand_num1 = np.random.multivariate_normal(mean, cov, 500)
print(rand_num1)
rand_num1.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
#plt.figure(figsize=(20,20))
plt.title('Two in one scatter plot')
plt.xlabel('x1')
plt.ylabel('x2')

plt.scatter(rand_num[:,0], rand_num[:,1],s=20, label='0', markers='o')

plt.scatter(rand_num1[:,0], rand_num1[:,1], s=20, label='1', marker='o')
plt.legend()
plt.show()
```

Problem 5

```python
#using concatenation
data_combine = np.concatenate((rand_num, rand_num1),axis=0)
print(data_combine)
data_combine.shape
```

```python
#using vstack
combined_data = np.vsatck((rand_num, rand_num1))
combined_data.shape
```

```python
#Creating a (1000,3) ndarray
zeros = np.zeros(500)
ones = np.ones(500)
new_column = np.concatenate((zeros, ones))
new_ndarr = np.append(combined_data, new_column[:,None], 1)
print(new_ndarr)
new_ndarr.shape
```
